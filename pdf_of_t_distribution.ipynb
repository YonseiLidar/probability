```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

x = np.linspace(-3,3)
for df in [1,3,5,8,50,100]:
    a = stats.t(df)
    y = a.pdf(x)
    plt.plot(x,y,label='Degree of Freedom : {}'.format(df))

a = stats.norm()
y = a.pdf(x)
plt.plot(x,y,'-*',label='Normal')

plt.legend()
plt.show()
```
