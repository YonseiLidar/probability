# PDF of Normal Distribution

$$
f(x)=\frac{1}{\sqrt{2\pi\sigma^2}}exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)
$$

```python
# plot of standard normal pdf 

# import numpy as np for numerical computation 
import numpy as np

# import matplotlib.pyplot as plt for ploting
import matplotlib.pyplot as plt

# mean and standard deviation 
mu = 0 # mean 0
sigma = 1 # standard deviation 1

### compute pdf

# x values where we compute pdf
x = mu + sigma * np.linspace(-3,3,100) 

# compute pdf values
y = np.exp(-(x-mu)**2/(2*sigma**2)) / np.sqrt(2*np.pi*sigma**2) 

# plot pdf
plt.plot(x,y) # plot
plt.show() # show plot
```

```python
# plot of normal pdf with mean 0 and several different standard deviations 

# import numpy as np for numerical computation 
import numpy as np

# import matplotlib.pyplot as plt for ploting
import matplotlib.pyplot as plt

# mean and standard deviation 
mu = 0 # mean 0
sigmas = np.arange(1,5) # several different standard deviations 

### compute and plot pdf

# x values where we compute pdf
x = np.linspace(-12,12,100)

for sigma in sigmas:
    
    # compute pdf values
    y = np.exp(-(x-mu)**2/(2*(sigma**2)))/np.sqrt(2*np.pi*(sigma**2))
    
    # plot with label
    plt.plot(x, y, label='std {}'.format(str(sigma))) 

plt.legend() # show lable
plt.show() # show plot 
```

```python
# plot of normal pdf with several different means and standard deviation 1

# import numpy as np for numerical computation 
import numpy as np

# import matplotlib.pyplot as plt for ploting
import matplotlib.pyplot as plt

# mean and standard deviation 
mus = np.arange(-2,3) # several different means
sigma = 1 # standard deviation 1 

### compute and plot pdf

# x values where we compute pdf
x = np.linspace(-6,6,100)

for mu in mus:
    
    # compute pdf values
    y = np.exp(-(x-mu)**2/(2*(sigma**2)))/np.sqrt(2*np.pi*(sigma**2))
    
    # plot with label 
    plt.plot(x, y, label='mean {}'.format(str(mu))) 

plt.legend() # show lable 
plt.show() # show plot
```

# Sampling

```python
# sampling from standard normal distribution

# import numpy as np for numerical computation 
import numpy as np

# import matplotlib.pyplot as plt for ploting
import matplotlib.pyplot as plt

# fix seed so that random samples are not changing
np.random.seed(9)  

# sample size
sample_size = 1000 

# sampling from standard normal distribution
x = np.random.normal(0., 1., (sample_size,)) 

# plot histogram
plt.hist(x, bins=50) # plot histogram 
plt.title('standard normal') # put title
plt.ylabel('frequency') # put ylabel 
plt.grid(True) # put grid
plt.show() # show histogram
```

$$\begin{array}{lll}
\mbox{Standardization}&&X\sim N(\mu,\sigma^2)\quad\Rightarrow\quad \frac{X-\mu}{\sigma}\sim N(0,1^2)\\
\mbox{Reverse Standardization}&&Z\sim N(0,1^2)\quad\Rightarrow\quad Z*\sigma+\mu\sim N(\mu,\sigma^2)\\
\end{array}$$

```python
# sampling from normal distribution

# import numpy as np for numerical computation 
import numpy as np

# import matplotlib.pyplot as plt for ploting
import matplotlib.pyplot as plt

# fix seed so that random samples are not changing
np.random.seed(9)  

# sample size
sample_size = 1000 

# mean and standard deviation
mu = 0 # mean 0 
sigma = 2 # standard deviation 2

# sampling from normal distribution
x = np.random.normal(mu, sigma, (sample_size,)) 

# plot histogram
plt.hist(x, bins=50) # plot histogram 
plt.title('N({}, {})'.format(mu, sigma)) # put title 
plt.ylabel('frequency') # put ylabel  
plt.grid(True) # put grid 
plt.show() # show histogram
```

# Estimation of mu and sigma

Let $x_i$ be iid sample from $N(\mu,\sigma^2)$.
Then, $\mu$ and $\sigma^2$ can be estimated by $\hat{\mu}$ and $\hat{\sigma}^2$
where 
$$\begin{array}{lll}
\hat{\mu}&=&\frac{\sum_{i=1}^nx_i}{n}\\
\hat{\sigma}^2&=&\frac{\sum_{i=1}^n(x_i-\hat{\mu})^2}{n}
\end{array}$$

```python
# sampling from normal distributions with several different means

# import numpy as np for numerical computation 
import numpy as np

# import matplotlib.pyplot as plt for ploting
import matplotlib.pyplot as plt

# fix seed so that random samples are not changing
np.random.seed(9)  

# sample size
num_samples = 1000

# generate normal samples
x = np.random.normal(2, 2, (num_samples,))

# plot pdf of samples using plt.hist 
plt.hist(x, bins=25, color='red', alpha=0.5, density=True) 

# estimate mu and sigma from the sample
mu_hat = np.sum(x) / num_samples
sigma_square_hat = np.sum((x - mu_hat)**2) / num_samples
sigma_hat = np.sqrt(sigma_square_hat)

### plot pdf of normal distribution with estimated mu and sigma

# x values where we compute pdf
xp = mu_hat + sigma_hat * np.linspace(-3,3,100) 

# compute pdf values
yp = np.exp(-(xp-mu_hat)**2/(2*sigma_hat**2)) / np.sqrt(2*np.pi*sigma_hat**2) 

# plot pdf of normal distribution with estimated mu and sigma 
plt.plot(xp,yp,'--r')
plt.show()
```
