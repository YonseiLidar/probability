# 1D Simple Random Walk

$$
S_n=\sum_{i=1}^nX_i
$$
where $X_i$ are iid.
We set $S_0=0$ by definition.

```python
import numpy as np
import matplotlib.pyplot as plt

t_data = [0]
y_data = [0]
for i in range(1,100):
    # https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.randint.html
    # numpy.random.randint(low, high=None, size=None, dtype='l')
    # Return random integers from low (inclusive) to high (exclusive).
    t_data.append(i)
    y_data.append(y_data[-1] + np.random.randint(-10, 11))
    
plt.plot(t_data, y_data)
plt.show()
```

```python
# https://www.youtube.com/watch?v=udDod-Q1akg&list=PLXziV1DL41oj_WtNSq41Jm4msNZ9VO6H8&index=9&t=0s
# https://www.youtube.com/watch?v=vNh_cRbcqi4&list=PLXziV1DL41oj_WtNSq41Jm4msNZ9VO6H8&index=10&t=0s
# https://www.youtube.com/watch?v=7RXUAIXxiiY&list=PLXziV1DL41oj_WtNSq41Jm4msNZ9VO6H8&index=11&t=0s

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

dy = np.random.randint(-10, 11, size=(10,))
y = np.cumsum(dy)

print(dy)
print(y)

# https://docs.scipy.org/doc/numpy/reference/generated/numpy.insert.html
# numpy.insert(arr, obj, values, axis=None)[source]
# obj : int, slice or sequence of ints
# Object that defines the index or indices before which values is inserted.
# values : array_like
# Values to insert into arr.
y = np.insert(y, 0, 0)

print(y)
    
plt.plot(y, '-*')
plt.show()
```
