```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.random.exponential(scale=2., size=(10000,))
plt.hist(x, bins=100, density=True)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
u = np.random.uniform(low=0., high=3., size=(10000,))
x = 2*np.log(3/(3-u))
plt.hist(x, bins=100, density=True)
plt.show()
```
