```python
# U(-1/2,1/2) * U(-1/2,1/2)

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

n = 2 # S_n
NumSimu = 10000 # Number of simulations

x = np.random.uniform(-0.5, 0.5, (NumSimu,n))
A = np.sum(x, axis=1)

plt.hist(A, bins=100)
plt.show()
```

```python
def plot_hist_of_S_n(n):
    import numpy as np
    import matplotlib.pyplot as plt

    # n = 3 # S_n
    NumSimu = 10000 # Number of simulations

    x = np.random.uniform(-0.5, 0.5, (NumSimu,n))
    A = np.sum(x, axis=1)

    plt.hist(A, bins=100, density=True)
    plt.title("Histogram of S_{}".format(n))
    plt.show()
```

```python
plot_hist_of_S_n(1)
```

```python
plot_hist_of_S_n(2)
```

```python
plot_hist_of_S_n(3)
```

```python
for n in range(1,20):
    plot_hist_of_S_n(n)
```
