# Confidence Interval of p

$$\begin{array}{lll}
\mbox{$X_i$ iid $B(p)$}
&\Rightarrow&
\frac{S_n-np}{\sqrt{np(1-p)}}\approx N(0,1)\\
&\Rightarrow&
\frac{n\hat{p}-np}{\sqrt{n\hat{p}(1-\hat{p})}}\approx N(0,1)\\
&\Rightarrow&
-1.96\le \frac{n\hat{p}-np}{\sqrt{n\hat{p}(1-\hat{p})}}\le 1.96\\
&\Rightarrow&
\hat{p}-1.96\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}\le p \le \hat{p}+1.96\sqrt{\frac{\hat{p}(1-\hat{p})}{n}}\\
\end{array}$$
where
$$
\hat{p}=\frac{S_n}{n}
$$

```python
# Confidence Interval of p where X_i are iid B(p)

import numpy as np
import matplotlib.pyplot as plt

# B(p) 
n = 100 # number of random variables generate for each sample
p = 0.5
num_samples = 50
S_n = np.random.binomial(n, p, num_samples)

for i in range(num_samples):
    p_hat = S_n[i] / n
    left = p_hat - 1.96 * np.sqrt(p_hat*(1-p_hat)/n)
    right = p_hat + 1.96 * np.sqrt(p_hat*(1-p_hat)/n)
    plt.plot([left,right], [i,i], 'o-')
plt.plot([p,p], [0,num_samples-1], '--')
plt.show()
```

# Confidence Interval of mu

$$\begin{array}{lll}
\mbox{$X_i$ iid with mean $\mu$ and variance $\sigma^2$}
&\Rightarrow&
\frac{S_n-n\mu}{\sqrt{n\sigma^2}}\approx N(0,1)\\
&\Rightarrow&
\frac{n\hat{\mu}-n\mu}{\sqrt{n\hat{\sigma}^2}}\approx N(0,1)\\
&\Rightarrow&
-1.96\le \frac{n\hat{\mu}-n\mu}{\sqrt{n\hat{\sigma}^2}}\le 1.96\\
&\Rightarrow&
\hat{\mu}-1.96\sqrt{\frac{\hat{\sigma}^2}{n}}\le \mu \le \hat{\mu}+1.96\sqrt{\frac{\hat{\sigma}^2}{n}}\\
\end{array}$$
where
$$
\hat{\mu}=\frac{S_n}{n},\quad
\hat{\sigma}^2=\frac{\sum_{i=1}^n(X_i-\hat\mu)^2}{n}
$$

```python
# Confidence Interval of mu where X_i are iid U(a,b)

import numpy as np
import matplotlib.pyplot as plt

# U(a,b) 
a = 0.0
b = 1.0
n = 100 # number of random variables generate for each sample
num_samples = 50
samples = np.random.uniform(a, b, (n,num_samples))

for i in range(num_samples):
    sample = samples[:,i] 
    mu_hat = np.sum(sample) / n
    sigma_square_hat = np.sum((sample - mu_hat)**2) / n  
    left = mu_hat - 1.96 * np.sqrt(sigma_square_hat/n)
    right = mu_hat + 1.96 * np.sqrt(sigma_square_hat/n)
    plt.plot([left,right],[i,i],'o-')
plt.plot([(b-a)/2, (b-a)/2],[0,num_samples-1],'--')
plt.show()
```

### Example - 95\% confidence interval

An astronomer is interested in measuring the distance, in light-years, from his observatory to a distant star. 
Although the astronomer has a measuring technique, he knows that, because of changing atmospheric conditions and normal error, each time a measurement is made it will not yield the exact distance, but merely an estimate. 
As a result, the astronomer plans to make a series of measurements and then use the average value of these measurements as his estimated value of the actual distance. 
If the astronomer believes that the values of the measurements are independent and iically distributed random variables having a common mean $d$ (the actual distance) and a common variance of 4 (light-years), how many measurements need he make to be reasonably sure that his estimated distance is accurate to within $\pm 0.5$ light-year?

For $X_k$, iid with mean $d$ and variance 4
\begin{eqnarray}
\frac{\sum_{k=1}^nX_k-nd}{\sqrt{4n}}\approx N(0,1)
&\Rightarrow&
\bar X=\frac{\sum_{k=1}^nX_k}{n}\approx \sqrt{\frac{4}{n}}\cdot N(0,1)+d\nonumber\\
&\Rightarrow&
|\bar X-d|\le 1.96\sqrt{\frac{4}{n}}
\le 0.5
\quad
\mbox{with 95% confidence}\nonumber\\
&\Rightarrow&
n\ge 61.4656
\quad
\mbox{Need at least 62 measurements!}
\nonumber
\end{eqnarray}
