```python
def plot_hist_of_S_n(n):
    import numpy as np
    import matplotlib.pyplot as plt
    
    %matplotlib inline

    # n = 3 # S_n
    NumSimu = 10000 # Number of simulations

    # https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.exponential.html
    x = np.random.exponential(scale=1/0.5, size=(NumSimu,n)) # scale=1/lambda
    A = np.sum(x, axis=1)

    plt.hist(A, bins=100, density=True)
    plt.title("Histogram of S_{}".format(n))
    plt.show()
```

```python
for n in range(1,20):
    plot_hist_of_S_n(n)
```
