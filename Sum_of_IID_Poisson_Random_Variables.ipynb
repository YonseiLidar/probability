```python
def plot_hist_of_S_n(n):
    import numpy as np
    import matplotlib.pyplot as plt
    
    %matplotlib inline

    # n = 3 # S_n
    NumSimu = 10000 # Number of simulations

    # https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.poisson.html
    x = np.random.poisson(lam=0.5, size=(NumSimu,n)) 
    A = np.sum(x, axis=1)

    plt.hist(A, bins=100, density=True)
    plt.title("Histogram of S_{}".format(n))
    plt.show()
```

```python
for n in range(1,20):
    plot_hist_of_S_n(n)
```

```python
for n in range(100,1000,100):
    plot_hist_of_S_n(n)
```

```python
for n in range(1000,10000,1000):
    plot_hist_of_S_n(n)
```

```python
plot_hist_of_S_n(10000)
```
